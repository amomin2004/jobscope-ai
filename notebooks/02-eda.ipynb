# Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(data=df, x="experience_level", y="salary_usd", order=['EN', 'MI', 'SE', 'EX'])
plt.title("Average Salary by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Average Salary (USD)")

```

```python
sns.barplot(data=df, x="company_size", y="salary_usd", order=['S','M','L'])
plt.title("Avg Salary by Company Size")
plt.xlabel("Company Size")
plt.ylabel("Average Salary (USD)")
```

```python
sns.barplot(data=df, x='industry', y="salary_usd")
plt.title("Avg Salary by Industry")
plt.xlabel('Industry')
plt.ylabel('Average Salary (USD)')
plt.xticks(rotation=45, ha='right')
```

```python
sns.barplot(data=df, x="education_required", y="salary_usd")
plt.title("Average Salary by Education")
plt.xlabel("Education")
plt.ylabel("Average Salary (USD)")
```

```python
sns.barplot(data=df, x="years_experience", y="salary_usd")
plt.title("Average Salary by YOE")
plt.xlabel("YOE")
plt.ylabel("Average Salary (USD)")
```

```python
sns.barplot(data=df, x="job_title", y="salary_usd")
plt.title("Average Salary by Job Title")
plt.xlabel("Job Title")
plt.ylabel("Average Salary (USD)")
plt.xticks(rotation=45, ha='right')
```

```python
sns.barplot(data=df, x="employment_type", y="salary_usd")
plt.title("Average Salary by Employment Type")
plt.xlabel("Employment Type")
plt.ylabel("Average Salary (USD)")
```
